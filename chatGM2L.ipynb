```python
import requests
from bs4 import BeautifulSoup

url = 'https://lyrics.lyricfind.com/lyrics/malik-montana-robie-yeah'
soup = BeautifulSoup(requests.get(url).content, 'html.parser')

print(soup.prettify())
```

```python
with open("output1.html", "w") as file:
    file.write(str(soup))
```

```python
with open("output1.html") as file:
    soup = BeautifulSoup(file, 'lxml')

```

```python
import json
json_text = json.loads(soup.body.script.text)

print(json_text['props']['pageProps']['songData']['track']['lyrics'])

with open("output2.json", "w") as file:
    json.dump(json_text, file, indent=4)
```
